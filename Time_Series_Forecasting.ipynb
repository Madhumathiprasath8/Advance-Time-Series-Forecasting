```python
"""
time_series_attention_project.py

Complete pipeline:
- Synthetic multivariate time series generation (4 features + target, T=1500)
- Preprocessing (MinMaxScaler), sliding windows
- Models:
    * Attention LSTM (stacked LSTMs + self-attention) -> returns (forecast, attention_weights)
    * Simple LSTM baseline
    * ARIMA baseline (univariate)
- Evaluation: RMSE, MAE, MAPE (on original scale)
- Attention weights extraction and plotting
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")

# -----------------------------
# Utils
# -----------------------------
def mape(y_true, y_pred, eps=1e-8):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / (y_true + eps))) * 100.0

def create_supervised_windows(data, lookback, target_col):
    X, y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:i+lookback, :])
        y.append(data[i+lookback, target_col])
    return np.array(X), np.array(y)

# -----------------------------
# Synthetic data generation
# -----------------------------
def generate_synthetic_series(T=1500, seed=42):
    rng = np.random.RandomState(seed)
    dates = pd.date_range("2015-01-01", periods=T, freq="D")
    t = np.arange(T)

    temp_trend = 0.01 * t
    temp_season = 5 * np.sin(2 * np.pi * t / 365.0)
    feat_temp = 20 + temp_trend + temp_season + rng.normal(scale=0.5, size=T)

    marketing_base = 50 + 10 * np.sin(2 * np.pi * t / 7.0)
    feat_marketing = marketing_base + rng.normal(scale=5.0, size=T)

    econ_trend = 0.005 * t
    econ_cycle = 3 * np.sin(2 * np.pi * t / 730.0)
    feat_econ = 100 + econ_trend + econ_cycle + rng.normal(scale=0.3, size=T)

    feat_event = rng.poisson(0.2, size=T).astype(float) + rng.normal(scale=0.1, size=T)

    target_trend = 0.02 * t
    target_season = 10 * np.sin(2 * np.pi * t / 365.0 + 0.5)
    noise = rng.normal(scale=2.0, size=T)

    target_sales = (
        0.6 * feat_temp
        + 0.3 * feat_marketing
        + 0.1 * feat_econ
        + 3.0 * feat_event
        + target_trend
        + target_season
        + noise
    )

    df = pd.DataFrame({
        "feat_temp": feat_temp,
        "feat_marketing": feat_marketing,
        "feat_econ": feat_econ,
        "feat_event": feat_event,
        "target_sales": target_sales,
    }, index=dates)

    return df

# -----------------------------
# Self-attention layer (returns weights when requested)
# -----------------------------
class SelfAttention(tf.keras.layers.Layer):
    """
    Self-attention that can return context and attention weights.
    When used with return_attention=True, .call() returns (context, att_weights).
    """
    def __init__(self, return_attention=False, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.return_attention = return_attention
        self.W = None
        self.b = None

    def build(self, input_shape):
        hidden_dim = int(input_shape[-1])
        self.W = self.add_weight(
            shape=(hidden_dim, 1),
            initializer="glorot_uniform",
            trainable=True,
            name=self.name + "_W"
        )
        self.b = self.add_weight(
            shape=(1,),
            initializer="zeros",
            trainable=True,
            name=self.name + "_b"
        )
        super(SelfAttention, self).build(input_shape)

    def call(self, inputs):
        # inputs: (batch, time, hidden)
        scores = tf.tensordot(inputs, self.W, axes=[[2], [0]]) + self.b  # (batch, time, 1)
        scores = tf.nn.tanh(scores)
        weights = tf.nn.softmax(scores, axis=1)  # (batch, time, 1)
        context = tf.reduce_sum(inputs * weights, axis=1)  # (batch, hidden)
        weights_squeezed = tf.squeeze(weights, axis=-1)    # (batch, time)
        if self.return_attention:
            return context, weights_squeezed
        return context

    def get_config(self):
        cfg = super(SelfAttention, self).get_config()
        cfg.update({"return_attention": self.return_attention})
        return cfg

# -----------------------------
# Model builders
# -----------------------------
def build_attention_lstm_model_with_att(lookback, n_features):
    """
    Returns a model with two outputs: forecast and attention weights.
    Train only uses forecast output (we set the attention loss weight to 0).
    """
    inputs = Input(shape=(lookback, n_features), name="input_window")
    x = LSTM(64, return_sequences=True, name="lstm_1")(inputs)
    x = LSTM(32, return_sequences=True, name="lstm_2")(x)
    att_layer = SelfAttention(return_attention=True, name="self_attention")
    context, att_weights = att_layer(x)  # att_weights shape = (batch, lookback)
    x = Dense(32, activation="relu", name="dense_1")(context)
    forecast = Dense(1, activation="linear", name="forecast")(x)

    model = Model(inputs=inputs, outputs=[forecast, att_weights], name="LSTM_Attention_with_weights")
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss=["mse", "mse"],          # second loss dummy
        loss_weights=[1.0, 0.0],     # ignore attention weights in loss
        metrics={"forecast": ["mae"]}
    )
    return model

def build_simple_lstm_model(lookback, n_features):
    inputs = Input(shape=(lookback, n_features), name="input_window")
    x = LSTM(32, name="lstm")(inputs)
    output = Dense(1, activation="linear", name="forecast")(x)
    model = Model(inputs, output, name="Baseline_LSTM")
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss="mse",
                  metrics=["mae"])
    return model

# -----------------------------
# ARIMA baseline helper
# -----------------------------
def arima_forecast(train_series, test_series, order=(5, 1, 0)):
    history = list(train_series)
    preds = []
    for t in range(len(test_series)):
        model = ARIMA(history, order=order)
        model_fit = model.fit()
        yhat = model_fit.forecast(steps=1)[0]
        preds.append(yhat)
        history.append(test_series[t])
    return np.array(preds)

# -----------------------------
# Main pipeline
# -----------------------------
def main():
    print("üîÑ Generating synthetic series...")
    df = generate_synthetic_series(T=1500)
    print(df.head())
    print(df.describe())

    # quick plot (optional)
    plt.figure()
    df["target_sales"].plot(title="Synthetic target_sales (first view)")
    plt.xlabel("Date"); plt.ylabel("Sales")
    plt.tight_layout(); plt.show()

    n_total = len(df)
    n_train = int(n_total * 0.7)
    n_val = int(n_total * 0.15)
    n_test = n_total - n_train - n_val

    train_df = df.iloc[:n_train]
    val_df = df.iloc[n_train:n_train + n_val]
    test_df = df.iloc[n_train + n_val:]

    print(f"\nSplit: train={len(train_df)}, val={len(val_df)}, test={len(test_df)}")

    # Scale using train stats
    scaler = MinMaxScaler()
    scaler.fit(train_df.values)
    train_scaled = scaler.transform(train_df.values)
    val_scaled = scaler.transform(val_df.values)
    test_scaled = scaler.transform(test_df.values)

    n_features = train_scaled.shape[1]
    target_col = df.columns.get_loc("target_sales")

    lookback = 30

    X_train, y_train = create_supervised_windows(train_scaled, lookback, target_col)
    X_val, y_val = create_supervised_windows(val_scaled, lookback, target_col)
    X_test, y_test = create_supervised_windows(test_scaled, lookback, target_col)

    print(f"\nWindows shapes: X_train={X_train.shape}, X_val={X_val.shape}, X_test={X_test.shape}")

    # Build models
    att_model = build_attention_lstm_model_with_att(lookback, n_features)
    att_model.summary()

    base_model = build_simple_lstm_model(lookback, n_features)
    base_model.summary()

    es = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

    # For training, we need a dummy attention target (zeros) since model has two outputs.
    zeros_train = np.zeros((len(y_train), lookback))
    zeros_val = np.zeros((len(y_val), lookback))

    print("\nüöÄ Training Attention LSTM (this may take ~ a few minutes)...")
    history_att = att_model.fit(
        X_train, [y_train, zeros_train],
        validation_data=(X_val, [y_val, zeros_val]),
        epochs=50, batch_size=32, callbacks=[es], verbose=1
    )

    print("\nüöÄ Training Baseline LSTM...")
    history_base = base_model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=50, batch_size=32, callbacks=[es], verbose=1
    )

    # ARIMA baseline on original unscaled target
    print("\nüìà Running ARIMA baseline (univariate on original target) ...")
    arima_train = train_df["target_sales"].values
    arima_test = test_df["target_sales"].values
    arima_preds = arima_forecast(arima_train, arima_test, order=(5,1,0))

    # Evaluate deep models: predict returns [forecast_preds, att_weights]
    print("\nüîé Predicting on test set...")
    preds_att, att_weights_test = att_model.predict(X_test, verbose=0)  # preds_att shape (N,1), att_weights (N, lookback)
    preds_att = preds_att.flatten()
    preds_base = base_model.predict(X_test, verbose=0).flatten()

    # Inverse-scaling helper (applies scaler inverse but only for target column)
    def inverse_target(scaled_arr):
        dummy = np.zeros((len(scaled_arr), n_features))
        dummy[:, target_col] = scaled_arr
        inv = scaler.inverse_transform(dummy)
        return inv[:, target_col]

    y_test_orig = inverse_target(y_test)
    preds_att_orig = inverse_target(preds_att)
    preds_base_orig = inverse_target(preds_base)

    # Metrics
    att_rmse = np.sqrt(mean_squared_error(y_test_orig, preds_att_orig))
    att_mae = mean_absolute_error(y_test_orig, preds_att_orig)
    att_mape = mape(y_test_orig, preds_att_orig)

    base_rmse = np.sqrt(mean_squared_error(y_test_orig, preds_base_orig))
    base_mae = mean_absolute_error(y_test_orig, preds_base_orig)
    base_mape = mape(y_test_orig, preds_base_orig)

    # ARIMA eval alignment: remove first 'lookback' samples from test for fair comparison
    if len(arima_preds) > lookback:
        arima_eval_true = arima_test[lookback:]
        arima_eval_pred = arima_preds[lookback:lookback + len(arima_eval_true)]
        arima_rmse = np.sqrt(mean_squared_error(arima_eval_true, arima_eval_pred))
        arima_mae = mean_absolute_error(arima_eval_true, arima_eval_pred)
        arima_mape = mape(arima_eval_true, arima_eval_pred)
    else:
        arima_rmse = arima_mae = arima_mape = np.nan

    print("\n===== TEST METRICS (original scale) =====")
    print("Attention LSTM:")
    print(f"  RMSE = {att_rmse:.3f}")
    print(f"  MAE  = {att_mae:.3f}")
    print(f"  MAPE = {att_mape:.2f}%")
    print("\nSimple LSTM Baseline:")
    print(f"  RMSE = {base_rmse:.3f}")
    print(f"  MAE  = {base_mae:.3f}")
    print(f"  MAPE = {base_mape:.2f}%")
    print("\nARIMA Baseline (aligned):")
    print(f"  RMSE = {arima_rmse:.3f}")
    print(f"  MAE  = {arima_mae:.3f}")
    print(f"  MAPE = {arima_mape:.2f}%")

    # Plot a small slice of test predictions
    n_plot = min(200, len(y_test_orig))
    plt.figure()
    plt.plot(test_df.index[lookback:lookback+n_plot], y_test_orig[:n_plot], label="True")
    plt.plot(test_df.index[lookback:lookback+n_plot], preds_att_orig[:n_plot], label="Attention LSTM")
    plt.plot(test_df.index[lookback:lookback+n_plot], preds_base_orig[:n_plot], label="Simple LSTM")
    plt.title("Test predictions (slice)")
    plt.xlabel("Time"); plt.ylabel("Sales"); plt.legend(); plt.tight_layout(); plt.show()

    # Attention weights: available as numpy array att_weights_test (N, lookback)
    print("\nüîç Attention weights shape (test samples x lookback):", att_weights_test.shape)
    # Example: visualize attention for first test sample
    sample_idx = 0
    sample_att = att_weights_test[sample_idx]  # numpy array of length lookback
    plt.figure()
    # <-- FIX: remove use_line_collection argument (some matplotlib versions don't accept it)
    plt.stem(range(lookback), sample_att)
    plt.title("Attention weights for first test sample (0=oldest, last=most recent)")
    plt.xlabel("Time step in window"); plt.ylabel("Attention weight"); plt.tight_layout(); plt.show()

    # print top-k attended positions for that sample
    top_k = 5
    top_idx = np.argsort(sample_att)[-top_k:][::-1]
    print(f"\nTop {top_k} attended positions in the input window (0=oldest):")
    for pos in top_idx:
        print(f"  Step {pos}: weight={sample_att[pos]:.4f}")

    print("\nPipeline finished ‚úÖ. Use metrics and attention analysis in your report.")

if __name__ == "__main__":
    main()
```
